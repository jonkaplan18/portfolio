```
# load data
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO 
from IPython.display import Image 
from pydot import graph_from_dot_data
import pandas as pd
import numpy as np

# nom_awards=pd.read_csv('final_nominations_df.csv')
# nom_awards
nom_awards=pd.read_csv('https://www.dropbox.com/s/ysuekksts86beag/final_nominations_df.csv?dl=1') #this is updated version of nom awards with AgeNominated, PastNominations and gender

# missing values
def missing_values_table(df):
        mis_val = df.isnull().sum()
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        return mis_val_table_ren_columns

missing_values_table(nom_awards)
```

```
# filter dataset down to numeric  , 3030 rows now

logit_df=nom_awards[['bookID', 'num_pages', 'publication_year', 'male', 'female', 'Winner', 'author_followers', 'average_rating', 'ratings_count', 'text_reviews_count', 'goodreads_author', 'book_series', 'AwardCategory', 'AwardEntity',  'PastNominations', 'authorName', 'author_bio', 'gender', 'description', 'bookTitle', 'HugoAward', 'NebulaAward', 'LocusAward']]
logit_df.dropna(inplace=True)
logit_df
```

```
# see imbalanced data
logit_df.Winner.value_counts()
```

```
# balance the data

from sklearn.utils import resample



# separate minority and majority classes
lost = logit_df[logit_df.Winner==0]
won = logit_df[logit_df.Winner==1]
# upsample minority
win_resampled = resample(won,
                          replace=True, # sample with replacement
                          n_samples=len(lost), # match number in majority class
                          random_state=42) # reproducible results
# combine majority and upsampled minority
upsampled = pd.concat([lost,win_resampled])

# check new class counts
upsampled.Winner.value_counts()
```

```
# choose variables for decision tree

import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt 
plt.rc("font", size=14)
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
sns.set(style="white")
sns.set(style="whitegrid", color_codes=True)
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO 
from IPython.display import Image 
from pydot import graph_from_dot_data
import pandas as pd
import sklearn.metrics as metrics

cols=['average_rating', 'publication_year', 'ratings_count', 'text_reviews_count', 'PastNominations', 'book_series', 'author_followers', 'NebulaAward', 'HugoAward', 'LocusAward']
x= upsampled[cols]
y= upsampled['Winner']

# test and train
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=1)
```
